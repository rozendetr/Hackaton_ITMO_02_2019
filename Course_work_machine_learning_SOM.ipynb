```
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn
import sklearn.cluster
```

```
data = pd.read_csv('/content/drive/My Drive/Course work Machine Learning/gender_set_normal.csv', header=0, sep=',')
data.info()
```

```
data_array=np.array(data.values)
```

```
!pip install neupy 
```

```
from neupy import algorithms, layers, utils
from neupy.layers import *
utils.reproducible()
```

```
sofm = algorithms.SOFM(
        # Use only two features for the input
        n_inputs=8,

        # Number of outputs defines number of features
        # in the SOFM or in terms of clustering - number
        # of clusters
        n_outputs=2,

        # In clustering application we will prefer that
        # clusters will be updated independently from each
        # other. For this reason we set up learning radius
        # equal to zero
        learning_radius=0,

        # Training step size or learning rate
        step=0.25,

        # Shuffles dataset before every training epoch.
        shuffle_data=True,

        # Instead of generating random weights
        # (features / cluster centers) SOFM will sample
        # them from the data. Which means that after
        # initialization step 3 random data samples will
        # become cluster centers
        weight='sample_from_data',

        # Shows training progress in terminal
        verbose=True,
    )
```

```
sofm.train(data_array, epochs=200)
```

```
labels_predict=sofm.predict(data_array)
```

```
labels=[]
for i in labels_predict:
  if i[1]==1:
    labels.append(0)
  else:
    labels.append(1)
```

```
data['gender']=labels
```

```
data['gender'].hist(bins=50)
```

```
print('0 = ', len(data.loc[data['gender'] == 0])/len(data)*100, '%')
print('1 = ', len(data.loc[data['gender'] == 1])/len(data)*100, '%')
```

metrics

```
from sklearn import metrics
```

```
sil_coeff = sklearn.metrics.silhouette_score(data_array, labels, metric='euclidean')
dav_bould_score = sklearn.metrics.davies_bouldin_score(data_array, labels)


print('The Silhouette Coefficient is ',sil_coeff)
print('The Davies-Bouldin score is ',dav_bould_score)

```

```
from sklearn.decomposition import PCA
from sklearn.metrics.pairwise import cosine_similarity
```

```
dist = 1 - cosine_similarity(data_array)
```

```
pca = PCA(2)
pca.fit(dist)
X_PCA = pca.transform(dist)
X_PCA.shape
```

```
x, y = X_PCA[:, 0], X_PCA[:, 1]
```

```
colors = {0: 'red',
          1: 'blue' }
names = {0: 'men', 
         1: 'women' }
```

```
df = pd.DataFrame({'x': x, 'y':y, 'gender':labels}) 
groups = df.groupby('gender')
```

```
fig, ax = plt.subplots(figsize=(20, 13)) 

for name, group in groups:
    ax.plot(group.x, group.y, marker='o', linestyle='', ms=5,
            color=colors[name],label=names[name], mec='none')
    ax.set_aspect('auto')
    ax.tick_params(axis='x',which='both',bottom='off',top='off',labelbottom='off')
    ax.tick_params(axis= 'y',which='both',left='off',top='off',labelleft='off')
    
ax.legend()

plt.show()
```
